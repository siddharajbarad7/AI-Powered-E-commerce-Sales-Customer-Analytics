```python
# Remove rows missing CustomerID or Description
data_cleaned = data.dropna(subset=['Customer ID', 'Description'], inplace=False)
```

```python
# 1. Remove missing critical values
data_cleaned
```

```python

# Convert Customer ID to integer on cleaned data for cleaner analysis

data_cleaned.loc[:, 'Customer ID'] = data_cleaned['Customer ID'].astype(int)
```

```python
data_cleaned
```

```python
data.head()
```

```python
# Remove cancellations (InvoiceNo starts with 'C')\
data_cleaned = data_cleaned.loc[
    ~data_cleaned['Invoice'].astype(str).str.startswith('C')
].copy()
```

```python
data_cleaned
```

```python
# Remove returns and bad prices
data_cleaned = data_cleaned.loc[
    (data_cleaned['Quantity'] > 0) & (data_cleaned['Price'] > 0)
].copy()
```

```python
 # Feature Engineering like a find a revenue column
data_cleaned['Revenue'] = data_cleaned['Quantity'] * data_cleaned['Price']
```

```python
data_cleaned
```

```python
data_cleaned.head(10)
```

```python
data_cleaned['InvoiceDate'] = pd.to_datetime(
    data_cleaned['InvoiceDate'],
    format='%d-%m-%Y %H:%M'
)
```

```python
# (very useful for EDA)
data_cleaned['Month'] = data_cleaned['InvoiceDate'].dt.month
data_cleaned['Year'] = data_cleaned['InvoiceDate'].dt.year
data_cleaned['DayNumber'] = data_cleaned['InvoiceDate'].dt.weekday + 1
data_cleaned['DayOfWeek'] = data_cleaned['InvoiceDate'].dt.day_name()
```

```python

#  Final Summary
print("Clean shape:", data_cleaned.shape)
print("Unique customers:", data_cleaned['Customer ID'].nunique())
print("Total revenue: £{:,.2f}".format(data_cleaned['Revenue'].sum()))
```

```python
# 1. CUSTOMERS — unique customers
customers = (data_cleaned[['Customer ID','Country']]
             .drop_duplicates(subset=['Customer ID'])
             .reset_index(drop=True))
print("Unique customers:", customers.shape[0])
```

```python
# 2. PRODUCTS — unique products
products = (
    data_cleaned[['StockCode', 'Description', 'Price']]
    .drop_duplicates(subset=['StockCode'])
    .reset_index(drop=True)
)
print("Unique products:", products.shape[0])
```

```python
# 3. ORDERS — unique invoices
orders = (data_cleaned[['Invoice','Customer ID','InvoiceDate']]
          .drop_duplicates(subset=['Invoice'])
          .reset_index(drop=True))
print("Unique orders:", orders.shape[0])
```

```python
# 4. ORDER ITEMS — every line item
order_items = (data_cleaned[['Invoice', 'StockCode', 'Quantity', 'Price', 'Revenue']]
               .reset_index(drop=True))
print("Total order items:", order_items.shape[0])
```

```python
print("customers :", customers.shape)
print("products  :", products.shape)
print("orders    :", orders.shape)
print("order_items:", order_items.shape)
```

```python
import os
os.makedirs('../data/clean', exist_ok=True)

customers.to_csv('../data/clean/customers.csv',   index=False)
products.to_csv('../data/clean/products.csv',     index=False)
orders.to_csv('../data/clean/orders.csv',         index=False)
order_items.to_csv('../data/clean/order_items.csv', index=False)

print("✓ All 4 CSVs saved to data/clean/")
```
